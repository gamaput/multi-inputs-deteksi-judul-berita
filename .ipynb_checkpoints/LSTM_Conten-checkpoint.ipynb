```python
data['Combined'] = data['title'] + ' ' + data['content']
data.head
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from keras.layers import Dense, Dropout, LSTM, Embedding, Flatten
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
```

```python
label = LabelEncoder().fit_transform(data["label"])
label
```

```python
x_train, x_test, y_train, y_test = train_test_split(data['Combined'], label, test_size=0.3, random_state=42)
input_length=100
```

```python
tok = Tokenizer(oov_token='UNK', num_words=5000)
tok.fit_on_texts(x_train)
```

```python
tok.word_index
```

```python
x_train_seq = tok.texts_to_sequences(x_train)
x_test_seq = tok.texts_to_sequences(x_test)
```

```python
x_train
```

```python
x_train_seq[0]
```

```python
x_train_pad = pad_sequences(x_train_seq, maxlen=input_length, padding='post')
x_test_pad = pad_sequences(x_test_seq, maxlen=input_length, padding='post')
```

```python
x_train_pad[0]
```

```python
model = Sequential([
                    Embedding(1000, 50, input_length=input_length),
                    LSTM(128, activation='sigmoid', return_sequences=True, name='LSTM'),
                    LSTM(100, activation='sigmoid', return_sequences=True),
                    Dropout(0.5, name='Dropout'),
                    LSTM(64, activation='sigmoid', name='LSTM'),
                    Dense(16, activation='relu', name='Dense'),
                    Dense(1, activation='sigmoid', name='Output')
])
```

```python
model.summary()
```

```python
keras.utils.plot_model(model, show_shapes=True)
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
print('Fitting model...')
hist = model.fit(x_train_pad,y_train, batch_size=64, epochs=25, validation_split=0.2, verbose=1)
```

```python
model.evaluate(x_test_pad, y_test)
```

```python
output = model.predict(x_test_pad)
print(output.shape)
print(output)
```

```python
model = Sequential()
model.add(Embedding(1000, 50, input_length=input_length))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
print('Fitting model...')
hist = model.fit(x_train_pad, y_train, batch_size=64, epochs=10, validation_split=0.1, verbose=1)
```

```python
model.evaluate(x_test_pad, y_test)
```

```python
y_predicted = model.predict(x_test_pad)
y_predicted = y_predicted.flatten()
```

```python
import numpy as np

y_predicted = np.where(y_predicted > 0.5, 1, 0)
y_predicted
```

```python
from sklearn.metrics import confusion_matrix, classification_report

cm = confusion_matrix(y_test, y_predicted)
cm 
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
print(classification_report(y_test, y_predicted))
```
